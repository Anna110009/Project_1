```python
from scipy.stats import f_oneway  # Import f_oneway function from scipy.stats

# Group final worth by industry category
industry_groups = [df[df['category'] == category]['finalWorth'] for category in df['category'].unique()]

# Perform ANOVA
f_statistic, p_value = f_oneway(*industry_groups)

print("ANOVA Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

### Since the p-value is much smaller than a typical significance level of 0.05, we would reject the null hypothesis. This suggests that there are statistically significant differences in the final worth of billionaires across different industry categories. In other words, the final worth of billionaires does vary significantly based on their industry category.
